## Analytical Data Model

An analytical data model is a structured representation of business data designed to support reporting, analytics, and machine learning. Unlike operational databases, which are optimized for transactional processes, analytical data models prioritize data consistency, accessibility, and efficient querying for decision-making purposes.

In this project, the analytical data model serves as the foundation for the entire analytics workflow. It integrates the raw source tables into a relational structure that enables exploratory data analysis, feature engineering, predictive modeling, and business decision optimization while preserving the integrity of the original data.

The design follows several key principles:

- Preserve the raw source data without modification.
- Define clear relationships between business entities.
- Ensure data integrity through primary and foreign keys.
- Create reusable SQL views for analytical consumption.
- Separate the storage layer from the analytical layer.
- Provide a single, consistent source of truth for downstream analyses.

Rather than performing transformations directly within Python notebooks, the analytical data model centralizes data preparation within SQL. This approach improves reproducibility, scalability, and maintainability while closely reflecting how modern analytics platforms are designed in production environments.

The resulting model provides a robust foundation upon which all subsequent phases of the project are built, ensuring that analyses, machine learning models, and business recommendations are generated from a consistent and reliable data source.

## Analytical Modeling Decisions

The analytical data model was designed to support the complete lifecycle of the project, from data exploration and feature engineering to predictive modeling and business decision optimization. The design prioritizes reproducibility, scalability, and maintainability over short-term implementation convenience.

Several architectural decisions were made to ensure that the analytical layer accurately reflects real-world data engineering and analytics practices.

### Design Principles

The analytical data model follows the principles below:

- Preserve the raw source data without modification.
- Separate data storage from analytical consumption.
- Maintain normalized source tables where appropriate.
- Build reusable SQL views for downstream analysis.
- Centralize data transformations within the database.
- Provide a single, consistent source of truth for all analytical workflows.

### Design Decisions

| Decision | Rationale |
|----------|-----------|
| Preserve raw data | Ensures reproducibility and prevents accidental modification of the original source data. |
| Use a relational database | Maintains clear relationships between business entities while improving query performance and data integrity. |
| Perform transformations in SQL | Centralizes business logic, improves maintainability, and enables reusable analytical datasets. |
| Build analytical SQL views | Decouples raw data from downstream consumers such as Python notebooks and Power BI dashboards. |
| Keep Python focused on analytics | Python is reserved for exploratory analysis, feature engineering, machine learning, and optimization rather than data preparation. |
| Maintain a modular architecture | Each project component can evolve independently without impacting the overall analytical workflow. |

Collectively, these decisions establish a scalable analytical foundation that closely resembles the architecture commonly adopted in modern analytics engineering and data science environments.

## Table Design

The analytical database is composed of a collection of normalized tables, each representing a distinct business entity within the telecommunications domain. Each table has a well-defined business purpose and represents a single level of granularity. Collectively, these tables provide a comprehensive view of the customer lifecycle while enabling downstream analytical processes through relational joins.

## Primary & Foreign Keys

Primary and foreign keys define the relationships between business entities and ensure the integrity of the analytical data model. Together, they establish a consistent relational structure that enables accurate joins, prevents orphan records, and supports reliable analytical workflows.

The customer entity serves as the central table within the model. Supporting entities are linked to the customer through foreign key relationships, allowing customer information to be enriched with demographic, geographic, service, and business status attributes.

The key relationships are summarized below.

## Data Import Strategy

The data import process is designed to ensure that the original source data is ingested into the analytical database in a controlled, reproducible, and non-destructive manner.

Each source file is imported into its corresponding SQL table without modifying the underlying data. This approach preserves the integrity of the original dataset while establishing a stable foundation for downstream transformations and analytical processing.

The import workflow follows the sequence below:

1. Load raw Excel files from the project's `data/raw/` directory.
2. Create the analytical database.
3. Create relational tables with appropriate data types.
4. Import each source file into its corresponding table.
5. Validate row counts and primary key integrity.
6. Verify referential integrity across related tables.
7. Create analytical SQL views for downstream consumption.

The import process intentionally separates **data ingestion** from **data transformation**. Raw data is stored exactly as received, while all business logic and analytical transformations are implemented through SQL views and subsequent analytical workflows.

This strategy provides several advantages:

- Preserves the original source data.
- Ensures a reproducible ingestion process.
- Simplifies troubleshooting and data validation.
- Supports incremental improvements to the analytical layer.
- Establishes a consistent foundation for Python, Power BI, and machine learning workflows.

> **Implementation**
>
> The complete import process is implemented through dedicated SQL scripts located in the project's `/sql` directory, allowing the analytical database to be recreated from scratch at any time.

## SQL Views

Rather than querying raw database tables directly, all downstream analytical workflows—including exploratory data analysis, feature engineering, machine learning, and Power BI reporting—consume data through analytical views. This approach centralizes business logic within the database and ensures consistency across every stage of the project.

The analytical views are designed to:

- Encapsulate business logic in a single location.
- Simplify analytical queries.
- Eliminate repetitive SQL joins.
- Provide reusable datasets for multiple consumers.
- Separate the storage layer from the analytical layer.
- Improve maintainability and reproducibility.

The primary analytical views are summarized below.

| View | Purpose |
|------|---------|
| `analytics.vw_customer_analytics` | Integrates all customer entities into a single analytical dataset for exploratory analysis and reporting. |

The first analytical view serves as the primary interface between SQL and Python. Instead of joining multiple normalized tables within notebooks, Python consumes a single, well-defined analytical dataset optimized for downstream analysis.

The analytical layer follows a modular design in which additional views can be introduced as business requirements evolve without modifying the underlying source tables.
