## Business Problem

A telecommunications company is experiencing customer churn while operating under a limited customer retention budget. Although historical customer data is available, retention efforts are not optimized to maximize business value.

The challenge is not simply to identify customers who are likely to churn, but to determine which customers should be prioritized for retention initiatives in order to maximize the expected return on investment (ROI).

Before predictive models can be developed, it is essential to understand the available data, assess its quality, and determine whether it contains the information required to support data-driven retention decisions.

This notebook represents the first stage of that process by establishing a thorough understanding of the dataset and its suitability for solving the business problem.
